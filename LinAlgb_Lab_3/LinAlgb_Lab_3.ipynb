$$X = c\cdot \begin{bmatrix} 3\\6 \\\end{bmatrix}, Y = c\cdot \begin{bmatrix} 9\\12 \\\end{bmatrix}$$

$$\vec{X} = \begin{bmatrix} 3\\6 \end{bmatrix},  \vec{Y} = \begin{bmatrix} 9\\12 \\\end{bmatrix} $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X =np.array([3,6])
Y = np.array([9,12])
```

```python
c = np.arange(-20,20,0.5)
plt.scatter(c*X[0],c*X[1],color = "green", label = " vector X")
plt.scatter(c*Y[0],c*Y[1],color = "cyan",marker = "*", label = " vector Y")
plt.xlim(-20,20)
plt.ylim(-20,20)

plt.axhline(y=0, color='blue')
plt.axvline(x=0, color='red')

plt.title("Linear combination using different scalar values ", fontsize = 12)
plt.legend(loc='lower right')

plt.grid()
plt.show()
```

### First Trial of Span

$$S = \begin{Bmatrix} c_1 \cdot\begin{bmatrix} 17\\7 \\\end{bmatrix} + 
c_2 \cdot \begin{bmatrix} 3\\-9 \\\end{bmatrix}\end{Bmatrix} $$



$$\vec{X} = \begin{bmatrix} 17\\7 \end{bmatrix} +\vec{Y} = \begin{bmatrix} 3\\9 \\\end{bmatrix} $$

```python
vectorX = np.array([17,7])
vectorY= np.array([3,-9])
```

```python
R = np.arange(-20,20,2)
c1, c2 = np.meshgrid(R,R)
spanRx = c1*vectorX[0] + c2*vectorY[0]
spanRy = c1*vectorX[1] + c2*vectorY[1]
plt.scatter(spanRx,spanRy, s=20, alpha=0.75, color = "magenta",marker = "X", label = "point values")

plt.axhline(y=0, color='blue')
plt.axvline(x=0, color='red')

plt.title('First Trial of Span of 2 Vectors', fontsize=12)
plt.xlabel('x-axis', fontsize=12)
plt.ylabel('y-axis', fontsize=12)
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
